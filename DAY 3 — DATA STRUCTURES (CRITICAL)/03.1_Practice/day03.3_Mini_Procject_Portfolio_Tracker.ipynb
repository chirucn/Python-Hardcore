```python
#Mini Project: Portfolio Tracker
trades=[]
while True:
    Symbol=input("Enter Symbol:")
    Buy_price=float(input("Enter Entry Price:"))
    Exit=float(input("Enter Exit Price:"))
    Qty=int(input("Enter Quantity:"))

    trade= {
        "Symbol":Symbol,
        "Buy_price":Buy_price,
        "Exit":Exit,
        "Qty":Qty
    }
    trades.append(trade)

    cont=input("Do you want to add another trade? (y/n):")
    if cont.lower() != 'y':
        break

#Calculations
total_investment=0
total_pnl=0

for trade in trades:
    investment=trade["Buy_price"]*trade["Qty"]
    pnl=(trade["Exit"]-trade["Buy_price"])*trade["Qty"]

    total_investment+=investment
    total_pnl+=pnl

#Output
print("\nT---Portfolio Summary:")
print("Total Innvestment:", total_investment)
print("Total PnL:", total_pnl)
```
