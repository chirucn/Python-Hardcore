```python
#1. Lists
prices=[100,101,110,105,107]

print()

#Accessin elements
print(prices[0])
print(prices[-2])

print()

#Modifyng elements
prices.append(120)
prices.remove(105)
print(prices)
print()

#Looping through lists
for p in prices:
    print(p)
```

```python
#2.List Logic

#sum of list
total=0
for p in prices:
    total+=p
print(total)

#max number finder
max_num=prices[0]
for p in prices:
    if p>max_num:
        max_num=p
print(max_num)
```

```python
#3.Dictionarties
trade={"Symbol":"NIFTY",
       "Price":25000,
       "Qty":65
}

#Accessing elements
print(trade["Symbol"])
print(trade["Price"])

#Modifying elements
trade["Price"]=25500

print(trade)
```

```python
for key, value in trade.items():
    print(key, value)
```

```python
#5.List of Discussions (Real Quant Structures)
trades=[
    {"Symbol":"NIFTY", "Price":25000, "Qty":65},
    {"Symbol":"BANKNIFTY", "Price":35000, "Qty":35}
]

#Accesing
for t in trades:
    print(t["Symbol"], t["Price"], t["Qty"])
```
