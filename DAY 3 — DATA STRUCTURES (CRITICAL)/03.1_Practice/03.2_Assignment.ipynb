```python
#A-1: Average Calculator
def average(prices):
    return sum(prices)/len(prices)

prices = [19500, 19600, 19700]
print(average(prices))
```

```python
#A-2: Minimum Number finder
def mini_num(prices):
    min_num=prices[0]
    for p in prices:
        if p<min_num:
            min_num=p
    return min_num

prices=[100, 101,98]
mini_num(prices)
```

```python
#A-3: Count Even Numbers in aa list
def count_even(nums):
    count=0
    for i in nums:
        if i%2==0:
            count+=1
    return count

nums=[1,2,3,4,5,6]
count_even(nums)
```

```python
#A-4: Trade Logger
trades4 = []

while True:
    Symbol= input("Enter Symbol:")
    Price= float(input("Enter Price:"))
    Qty=int(input("Enter Qty:"))

    t1= {"Symbol":Symbol, "Price":Price, "Qty":Qty}
    trades4.append(t1)

    print("Trades added!")

    cont=input("Do you want to add another trade? (y/n):")
    cont.lower() != 'y'
    break

print("\nAll Trades:")
for t1 in trades4:
    print(t1)
```

```python
#A-5: Trade PnL Calculator
def trade_pnl(trades3):
    total_pnl = 0
    for trade in trades3:
        total_pnl += (trade["Exit"] - trade["Entry"]) * trade["Qty"]
    return total_pnl

trades3=[
    {"Entry": 25000, "Exit": 25100, "Qty": 65},
    {"Entry": 35000, "Exit": 34900, "Qty": 35}
]
trade_pnl(trades3)
```
