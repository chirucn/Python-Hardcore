```python
#1.Functions (most Important Concept)
def greet(name):
    print("Hello,", name)

greet("Chiranjeevi")

#with return
def add(a,b):
    return a+b
result=add(4,3)
print(result)

add(10,20)
```

```python
#3.Build Logic with Functions
#Even/Odd using functions
def check_even(num):
    if num%2==0:
        return "Even"
    else:
        return "Odd"

print(check_even(24))
```

```python
def table(n):
    for i in  range(1,11):
        print(n,"x",i,"=",n*i)

table(11)
```
