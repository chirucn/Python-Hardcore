```python
#A-1.Calculating using functions
def add(a,b):
    return a+b
print(add(4,3))

def subtract(a,b):
    return a-b
print(subtract(4,3))

def multiply(a,b):
    return a*b
print(multiply(4,3))    

def divide(a,b):
    return a/b
print(divide(4,3))
```

```python
#A-2:Prime Number Checker
def is_prime(n):
    if n<=1:
        return False
    for i in range(2,n):
        if n%i==0:
            return False

    return True

print(is_prime(17))
```

```python
#A-3:Factorial function
def factorial(x):
    if x<=0:
        return None
    
    result=1
    for i in  range(1,x+1):
        result*=i

    return result
print(factorial(10))
```

```python
#A-4: Sum of list
def sum_list(lst):
    total_sum=0
    for i in range(len(lst)):
        total_sum+=lst[i]
    return total_sum
print(sum_list([1,2,3,4,5]))
```

```python
#A-5: Max Number Finder in list
def max_in_list(lst):
    if len(lst)==0:
        return None
    max_num=lst[0]
    for i in range(len(lst)):
        if lst[i]>max_num:
            max_num=lst[i]
    return max_num
print(max_in_list([1,2,3,4,5]))
```
