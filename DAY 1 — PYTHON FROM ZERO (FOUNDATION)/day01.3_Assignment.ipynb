```python
# A-1: Square of a number
a=int(input("Enter a number:"))
print("Square of",a,"is",a*a)
```

```python
#A-2:Check Positive/Negative Number
x=float(input("Enter a number:"))
if x>=0:
    print("Positive")
else:
    print("Negative")
```

```python
#A-3: Print numbers 1-100 using loop
for i in range(1,101):
    print(i)
```

```python
#A-4:Sum of first N numbers
n=int(input("Enter a number:"))
total=0
for i in range(1,n+1):
    total+=i
print("Sum of first",n,"numbers is",total)
```

```python
#A-5: Multiplication Table(Very Important)
num=int(input("Enter a number:"))
for i in range(1,11):
    print(num,"x",i,"=",num*i)
```
