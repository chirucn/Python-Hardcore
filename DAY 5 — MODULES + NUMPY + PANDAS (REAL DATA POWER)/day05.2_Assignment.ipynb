👉 Add column pnl  
👉 Compute total PnL

```python
#A-3: PnL Calculator
df["PnL"]=(df["Sell_Price"]-df["Buy_Price"])*df["Qty"]
total_PnL=df["PnL"].sum()
print(df)
print("\nTotal PnL:",total_PnL)
```

**4. Filter Data**  
Show trades where profit > 0

```python
#A-4: Filter Data
Profit_trades=df[df["PnL"]>0]
print(Profit_trades)
```
