✅ Build: Trade Analysis Engine Features: 
    Read CSV  
    Compute PnL Show:  
        total profit  
        average trade  
        winning trades 

```python
# Mini Projeect: Trade Analysis Engine
import pandas as pd

#Read CSV 
df=pd.read_csv("trade_log.csv")

#Profit/Loss column
df["PnL"]=(df["Sell_Price"]-df["Buy_Price"])*df["Qty"]

#Total Pnl
total_pnl= df["PnL"].sum()

#Average Trade Profit
avg_trade_pnl= df["PnL"].mean()

#Winning Trades
winning_trades= df[df["PnL"]>0]
win_rate= (len(winning_trades)/len(df["PnL"]))*100

#Output
print("\n---- TRADE ANALYSIS ----")
print("Total P/L:", total_pnl)
print("Average Trade P/L:", avg_trade_pnl)
print("Winning Trades (%):", win_rate)
```
