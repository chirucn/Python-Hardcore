```python
#1.Basic Plot
import matplotlib.pyplot as plt

prices=[100,101,105,107,98,110]

plt.plot(prices)
plt.show()
```

```python
#3.Add Labels
plt.plot(prices)

plt.title("Price Movement")
plt.xlabel("Time")
plt.ylabel("Price")

plt.show()
```

```python
#4.Plot from CSV
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("trade_log.csv")

plt.plot(df["Buy_Price"])
plt.show()
```

```python
#5.Multiple Lines
plt.plot(df["Buy_Price"], label="Entry")
plt.plot(df["Sell_Price"], label="Exit")

plt.legend()
plt.show()
```

```python
#6.Bar Chart (PnL View)
df["PnL"]=(df["Sell_Price"]-df["Buy_Price"])*df["Qty"]
plt.bar(range(len(df)), df["PnL"])
plt.show()
```

```python
#7.Cummulative Pnl
df["Cum_pnl"]=df["PnL"].cumsum()

plt.plot(df["Cum_pnl"])
plt.title("Cummulative PnL")
plt.show()
```
