```python
# Mini Project: Strategy Visualizer
import pandas as pd
import matplotlib.pyplot as plt

#Load trade data
df = pd.read_csv("trade_log.csv")

#Computation of PnL
df["PnL"] = (df["Sell_Price"] - df["Buy_Price"]) * df["Qty"]

#Cumulative PnL (Equity Curve)
df["Cumulative_PnL"] = df["PnL"].cumsum()

#Price Trend
plt.figure()
plt.plot(df["Sell_Price"])
plt.title("Price Trend")
plt.xlabel("Trade Index")
plt.ylabel("Price")
plt.show()

#Trade PnL
plt.figure()
plt.bar(range(len(df)), df["PnL"])
plt.title("Trade PnL")
plt.xlabel("Trade Number")
plt.ylabel("PnL")
plt.show()


#Cumulative PnL (Equity Curve)

plt.figure()
plt.plot(df["Cumulative_PnL"])
plt.title("Equity Curve")
plt.xlabel("Trades")
plt.ylabel("Cumulative PnL")
plt.show()
```
