**1. Price Chart**  
Plot price from CSV


```python
#A-1: Price Chart
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv("trade_log.csv")

plt.plot(df["Buy_Price"])
plt.title("Price")
plt.show()
```

**2. PnL Chart**  
Plot individual trade PnL


```python
#A-2: PnL Chart
df["PnL"]=(df["Sell_Price"]-df["Buy_Price"])*df["Qty"]

plt.bar(range(len(df)), df["PnL"])
plt.title("Trade PnL")
plt.xlabel("Trade Number")
plt.ylabel("PnL")
plt.show()
```

**3. Cumulative PnL**  
Build equity curve


```python
#A-3: Cummulative PnL
df["Cum_PnL"]=df["PnL"].cumsum()

plt.plot(df["Cum_PnL"])
plt.title("Equity Curve")
plt.xlabel("Trades")
plt.ylabel("Cumulative PnL")
plt.show()
```

**4. Compare Entry vs Exit**  
Two-line chart


```python
plt.plot(df["Buy_Price"], label="Entry")
plt.plot(df["Sell_Price"], label="Exit")

plt.title("Entry vs Exit Prices")
plt.xlabel("Trades")
plt.ylabel("Price")
plt.legend()
plt.show()
```

**5. Winning vs Losing Trades**  
Count and visualize

```python
wins = len(df[df["PnL"] > 0])
losses = len(df[df["PnL"] <= 0])

plt.bar(["Wins", "Losses"], [wins, losses])
plt.title("Winning vs Losing Trades")
plt.show()
```
