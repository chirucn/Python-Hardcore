```python
#2.Basic class
class Trade:
    def __init__(self, symbol, price):
        self.symbol= symbol
        self.price= price


# Create Object
t1= Trade("NIFTY",24500)
print(t1.symbol)
print(t1.price)
```

```python
#3.Add Methods (Behavior)
class Trades:
    def __init__(self, symbol, entry, exit, qty):
        self.symbol = symbol
        self.entry = entry
        self.exit = exit
        self.qty = qty

    def pnl(self):
        return (self.exit-self.entry)*self.qty

```

```python
t1= Trades("NIFTY", 25000, 25200, 10)
print(t1.pnl())
```

```python
#5.Multiple Objects
trades = []

trades.append(Trades("NIFTY", 100, 110, 5))
trades.append(Trades("BANKNIFTY", 200, 190, 5))

for t in trades:
    print(t.symbol, t.pnl())
```

```python
#6.Class for Portfolio
class Portfolio:
    def __init__(self):
        self.trades = []

    def add_trade(self, trade):
        self.trades.append(trade)

    def total_pnl(self):
        total = 0
        for t in self.trades:
            total += t.pnl()
        return total
```
