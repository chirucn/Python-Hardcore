```python
#6.Working with CSV

# Write CSV
import csv

with open("trades.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    writer.writerow(["Symbol", "Price", "Quantity"])
    writer.writerow(["AAPL", 150, 10])
```

```python
# Read CSV
import csv

with open ("trades.csv", "r") as file:
    reader = csv.reader(file)
    
    for row in reader:
        print(row)
```

```python
#7.Real Trading Logger

import csv

Symbol= input("Enter Symbol:")
Price= float(input("Enter Price:"))

with open ("trades.csv", "a", newline="") as file:
    writer = csv.writer(file)
    
    writer.writerow([Symbol, Price])
```
