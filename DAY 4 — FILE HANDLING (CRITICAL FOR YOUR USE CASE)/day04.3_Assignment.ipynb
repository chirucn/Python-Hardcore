**1. File Reader**  
  Read a file  
  Count number of lines

```python
#A-1: File Reader
count=0
with open("data.txt", "r") as file:
    for line in file:
        count +=1

print(count)
```

**2. Word Counter**  
Count total words in file

```python
#A-2: Word Counter
word_count=0
with open("data.txt", "r") as file:
    for  line in file:
        words=line.split()
        word_count += len(words)

print(word_count)
```

**4. Trade Logger (UPGRADE)**  
Store:  
symbol  
price  
timestamp (use datetime)

```python
#A-4: Trade Logger
from datetime import datetime
import csv
import os

trade_log=[]

while True:
    Symbol=input("Enter the stock symbol: ")
    Buy_Price=float(input("Enter the buy price: "))
    Sell_Price=float(input("Enter the sell price: "))
    Qty=int(input("Enter the quantity: "))
    timestamp=datetime.now().strftime("%Y-%m-%d %H:%M:%S")

    t={"timestamp": timestamp,
       "Symbol": Symbol,
       "Buy_Price": Buy_Price,
       "Sell_Price": Sell_Price,
       "Qty": Qty}
    
    trade_log.append(t)

    cont=input("Do you want to add another trade? (y/n): ")
    if cont.lower() != 'y':
        break

#check if file exists
file_exists= os.path.exists("trade_log.csv")

with open("trade_log.csv", "a", newline="") as file:
    writer = csv.writer(file)

    if not file_exists:
        writer.writerow(["timestamp", "Symbol", "Buy_Price", "Sell_Price", "Qty"])

    for t in trade_log:
        writer.writerow([t["timestamp"], t["Symbol"], t["Buy_Price"], t["Sell_Price"], t["Qty"]])
    
```

**5. PnL from CSV**  
Read:  
entry,exit  
100,110  
200,190

```python
#A-5: PnL from CSV
import csv

with open("trade_log.csv", "r") as file:
    reader=csv.DictReader(file)

    total_Pnl=0
    for row in reader:
        try:
            PnL = (float(row["Sell_Price"]) - float(row["Buy_Price"])) * int(row["Qty"])
        except (ValueError, KeyError):
            continue
        total_Pnl += PnL
print("Total PnL:", total_Pnl)
```
