```python
#2.Reading Files
file = open("data.txt", "r")
content=file.read()
print(content)

file.close()
```

```python
# Better way of reading files
with open ("data.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
#3.Read Line by Line
with open ("data.txt", "r") as file:
    for line in file:
        print(line.strip())
        
```

```python
#4.Writing files
with open ("data.txt", "w") as file:
    file.write("NIFTY 26000\n")    

## This will overwrite the existing content of the file.
# If you want to append to the file, use "a" mode instead of "w".
```

```python
#5.Appennd(Very Important)
with open ("data.txt", "a") as file:
    file.write("BANKNIFTY 35000\n")

#Adds new data (used in logging systems)
```
